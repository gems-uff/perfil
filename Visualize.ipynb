```python
import bibliometrics
bibliometrics.setup(horizon=3)

subject = {
    'Nome': 'Leonardo Gresta Paulino Murta',
    'ID Lattes': '1565296529736448',
    'ID Scholar': 'VEbJeB8AAAAJ'
}

# Uncomment this line to add a red dot indicating the performance of the subject above in the following boxplot
# subject.update(bibliometrics.all(subject['ID Lattes'], subject['ID Scholar']))
```

```python
import pandas as pd
df = pd.read_csv('professores.csv', sep='\t', decimal=',')
```

```python
# Métricas referentes a toda a carreira
legends = ['Projetos', 'Projetos\nCoordenados', 'Orientações', 'Orientações\nde Doutorado', 'Bancas', 'Bancas\nde Doutorado', 'Publicações', 'Publicações\nem Periódicos', 'Publicações\nIndexadas JCR', 'Citações', 'H-Index']
metrics = [e.replace('\n', ' ') + ' (total)' for e in legends]
bibliometrics.boxplot(df, subject, metrics, legends, rows=2)
```
